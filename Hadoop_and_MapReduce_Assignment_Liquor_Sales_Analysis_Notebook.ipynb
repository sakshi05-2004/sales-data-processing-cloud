### Plotting box plots to identify outliers

```python
# Select only numerical columns
numerical_cols = df1.select_dtypes(include=['float64', 'int64']).columns

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Loop through each numerical column to create boxplots
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(len(numerical_cols) // 3 + 1, 3, i)  # Arrange in multiple rows and 3 columns
    sns.boxplot(x=df1[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()  # Adjust layout to prevent overlap

plt.show()
```

#### 3.1 Total Revenue by Store
<font color = red>[10 marks]</font> <br>

```python
from mrjob.job import MRJob
import csv
from io import StringIO

class MRTotalRevenueByStore(MRJob):

    def mapper(self, _, line):
        try:
            row = next(csv.reader(StringIO(line)))
            if row[0].strip() != 'Invoice/Item Number':
                store_number = row[2].strip()
                sale_dollars = float(row[21]) if row[21] else 0.0
                yield store_number, sale_dollars
        except Exception:
            pass

    def reducer(self, store_number, sale_dollars_list):
        yield store_number, sum(sale_dollars_list)

if __name__ == '__main__':
    MRTotalRevenueByStore.run()
```

#### 3.2 Top-Selling Categories
<font color = red>[10 marks]</font> <br>

```python
from mrjob.job import MRJob
import csv
from io import StringIO

class MRTopSellingLiquorCategories(MRJob):

    def mapper(self, _, line):
        try:
            reader = csv.reader(StringIO(line))
            row = next(reader)
            if row[0].strip() != 'Invoice/Item Number':
                category_name = row[11].strip()
                bottles_sold = int(row[20]) if row[20] else 0
                sale_dollars = float(row[21]) if row[21] else 0.0
                yield category_name, (bottles_sold, sale_dollars)
        except Exception:
            pass

    def reducer(self, category_name, values):
        total_bottles = 0
        total_sales = 0.0
        for bottles_sold, sale_dollars in values:
            total_bottles += bottles_sold
            total_sales += sale_dollars
        yield category_name, (total_bottles, total_sales)

if __name__ == '__main__':
    MRTopSellingLiquorCategories.run()
```

```python
from mrjob.job import MRJob
import csv
from io import StringIO

class MRCountyLevelSalesAnalysis(MRJob):

    def mapper(self, _, line):
        try:
            row = next(csv.reader(StringIO(line)))
            if row[0].strip() != 'Invoice/Item Number':
                county = row[9].strip()
                sale_dollars = float(row[21]) if row[21] else 0.0
                volume_litres = float(row[22]) if row[22] else 0.0
                volume_gallons = float(row[23]) if row[23] else 0.0
                yield county, (sale_dollars, volume_litres, volume_gallons)
        except Exception:
            pass

    def reducer(self, county, values):
        total_sales = 0.0
        total_litres = 0.0
        total_gallons = 0.0
        for sale_dollars, volume_litres, volume_gallons in values:
            total_sales += sale_dollars
            total_litres += volume_litres
            total_gallons += volume_gallons
        yield county, (total_sales, total_litres, total_gallons)

if __name__ == '__main__':
    MRCountyLevelSalesAnalysis.run()
```

#### 3.4 Store Performance Analysis
<font color = red>[10 marks]</font> <br>

```python
from mrjob.job import MRJob
import csv
from io import StringIO

class MRStorePerformanceAnalysis(MRJob):

    # Mapper: emit (Store Number, (Sale Dollars, Volume Litres, Transaction Count))
    def mapper(self, _, line):
        try:
            row = next(csv.reader(StringIO(line)))
            if row[0].strip() != 'Invoice/Item Number':
                store_number = row[2].strip()
                sale_dollars = float(row[21]) if row[21] else 0.0
                volume_litres = float(row[22]) if row[22] else 0.0
                yield store_number, (sale_dollars, volume_litres, 1)
        except Exception:
            pass  # Ignore parsing issues

    # Reducer: sum sale dollars, volume sold and transaction count per store
    def reducer(self, store_number, values):
        total_sales = 0.0
        total_volume = 0.0
        transaction_count = 0
        for sale_dollars, volume_litres, count in values:
            total_sales += sale_dollars
            total_volume += volume_litres
            transaction_count += count
        average_sale_per_transaction = total_sales / transaction_count if transaction_count != 0 else 0.0
        yield store_number, (total_sales, total_volume, average_sale_per_transaction)

if __name__ == '__main__':
    MRStorePerformanceAnalysis.run()
```

#### 3.5 Trends in Liquor Sales Over Time
<font color = red>[10 marks]</font> <br>

```python
from mrjob.job import MRJob
import csv
from io import StringIO

class MRLiquorSalesTrends(MRJob):

    # Mapper: emit (Year-Month, (Sale Dollars, Volume Sold Litres))
    def mapper(self, _, line):
        try:
            row = next(csv.reader(StringIO(line)))
            if row[0].strip() != 'Invoice/Item Number':  # Skip header
                date_str = row[1].strip()
                if date_str:
                    day, month, year = date_str.split('-')
                    year_month = f"{year}-{month.zfill(2)}"
                    sale_dollars = float(row[21]) if row[21] else 0.0
                    volume_sold_litres = float(row[22]) if row[22] else 0.0
                    yield year_month, (sale_dollars, volume_sold_litres)
        except Exception:
            pass  # Ignore parsing errors

    # Reducer: sum sale dollars and volume sold per year-month
    def reducer(self, year_month, values):
        total_sales = 0.0
        total_volume = 0.0
        for sale_dollars, volume_sold_litres in values:
            total_sales += sale_dollars
            total_volume += volume_sold_litres
        yield year_month, (total_sales, total_volume)

if __name__ == '__main__':
    MRLiquorSalesTrends.run()
```
